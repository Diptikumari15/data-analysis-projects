```python
# to change column_name
df.rename( columns = {"Pclass" : "Passenger_class"}, inplace=True)
```

```python
#to get sum of null rows in each column
pd.isnull(df).sum()
```

```python
#to fill null value in age
df['Age'].fillna(00, inplace=True)
```

```python
#to get the columns whose data type is object
df.dtypes[df.dtypes == object].index
```

```python
#to get analysis of columns whose data type is object
df[df.dtypes[df.dtypes == object].index].describe()
```

```python
#to get unique value of Passenger_class
pd.Categorical(df["Passenger_class"])
```

```python
# to get Name,Sex,Fare of people who are female and pays more than average Fare
df[(df["Sex"] == "female") & (df["Fare"] > 32)][["Name","Sex","Fare"]]
```

```python
# to get Name and Fare of people who pays more than maximum Fare
df[df["Fare"]  == max(df["Fare"])][["Name","Fare"]]
```

```python
g = df.groupby("Survived")
```

```python
#to group data on the basis of Survived and  get their statistical analysis
g.describe()
```

```python
#to get currency of Fair from Euro to INR
def euro_inr(x):
    return x*80

df["Fare_INR"] = df["Fare"].apply(euro_inr)
```

```python
df.head(2)
```

```python
# to categorize Fare into cheap, mid and high
def cat_fare(x):
    if x < 10:
        return "cheap"
    elif x>=10 and x<20:
        return "mid"
    else:
        return "high"
    
    
df["Cat_fare"] = df["Fare"].apply(cat_fare)    
    
```

```python
df.Cat_fare
```

```python
#to create a age range column
def age_r(x):
    if x < 25:
        return "0-25"
    elif x>=25 and x<50:
        return "25-50"
    else:
        return "50+"
    
    
df["Age_range"] = df["Age"].apply(age_r)    
    
```
